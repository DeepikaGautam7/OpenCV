```python
import cv2 as cv 
import matplotlib.pyplot as plt
import numpy as np
img = cv.imread('img/river.jpg')
plt.imshow(img)#default BGR
plt.show()
```

```python
res = (cv.cvtColor(img, cv.COLOR_BGR2GRAY))
plt.imshow(res, cmap = 'gray')
plt.show()
```

```python
res = (cv.cvtColor(img, cv.COLOR_BGR2RGB))
plt.imshow(res)
plt.show()
```

```python
res = (cv.cvtColor(img, cv.COLOR_BGR2HSV))
plt.imshow(res)
plt.show()
```
