```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
img = cv.imread('img/flower.jpg')
plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
plt.title("Original Image")
plt.show()
```

```python
x = np.ones(img.shape, dtype="uint8") * 60
#np.ones - create array of image size, img.shape - (h, w, channel), uint8 - px range (0, 255)  
add_img = cv.add(img, x)
plt.imshow(cv.cvtColor(add_img, cv.COLOR_BGR2RGB))
plt.title("Brightened Image ")
plt.show()
```

```python
x = np.ones(img.shape, dtype="uint8") * 50
#np.ones - create array of image size, img.shape - (h, w, channel), uint8 - px range (0, 255)  
plt.title("Darkened Image ")
sub_img = cv.subtract(img, x)
plt.imshow(cv.cvtColor(sub_img, cv.COLOR_BGR2RGB))
plt.show()
```

```python
mul_img = cv.multiply(img, np.array([2, 2, 2]))
plt.title("High Contrast Image")
plt.imshow(cv.cvtColor(mul_img, cv.COLOR_BGR2RGB))
plt.show()
```

```python
div_img = cv.divide(img, np.array([2,2,2]))
plt.title("Low Contrast Image")
plt.imshow(cv.cvtColor(div_img, cv.COLOR_BGR2RGB))
plt.show()
```
