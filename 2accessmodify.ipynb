```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv.imread('img/cat.jpg')
px = img[500,500]
print('Pixel at (150, 100):',px) #BRG
plt.title("Original Image")
plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
plt.show()
```

```python
img[500:550, 500:550] = [0, 0, 255] 
img[100:350, 50:250] = [255, 255, 255] 
plt.title("Modified Image")
plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
plt.show()
```
