```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv.imread('img/cat.jpg')
print('image(h,w,c)',img.shape)
plt.title("Original Image")
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

```python
w, h = 300, 300  
fixed = cv.resize(img, (w, h))

x, y = 0.75, 0.75  #scalefactor
scaled = cv.resize(img, None, fx=x, fy=y)

w = 800  
aspratio = img.shape[0] / img.shape[1]  # h / w
h = int(w * aspratio)
aspratio = cv.resize(img, (w, h))

plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.title("Fixed Size (300*300)")
plt.imshow(fixed)

plt.subplot(1, 3, 2)
plt.title("Scaled 75% (fx=0.75, fy=0.75)")
plt.imshow(scaled)

plt.subplot(1, 3, 3)
plt.title(f"Aspect Ratio Maintained ({w}x{h})")
plt.imshow(aspratio)
plt.show()
```
